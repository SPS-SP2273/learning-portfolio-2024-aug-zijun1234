```python
result=[]
for i in range(10):
    if i%2==0:
        continue
    result.append(i**2)
print(result)
```

```python
[i**2 for i in range(10) if i%2!=0]
```

```python
%%timeit
[i**2 for i in range(10) if i%2!=0]
#using list comprehension is much faster, since i am using print() with the for loop in the first method,
#error message occurs because %%timeit uses a large number of loops to measure the mean time,and print function
#print out every result leading in the error message.
```
