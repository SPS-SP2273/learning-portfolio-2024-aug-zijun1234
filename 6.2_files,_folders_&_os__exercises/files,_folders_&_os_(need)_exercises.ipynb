```python
import os
month_conversion={
'jan': '01', 'feb': '02', 'mar': '03',
'apr': '04', 'may': '05', 'jun': '06',
'jul': '07', 'aug': '08', 'sep': '09',
'oct': '10', 'nov': '11', 'dec': '12'
}
import glob
all_data=glob.glob('**/*.txt', recursive=True)
for files in all_data:
    old_file=files.split(os.path.sep)[-1]
    month_abbr=old_file.split('-')[0]
    month_number=month_conversion[month_abbr]
    new_file=old_file.replace(month_abbr,month_number)
    old_path=files
    new_path=os.path.join('.','os-collaboration-exercise-data',new_file)
    os.rename(old_path,new_path)
    print(f'{old_file} ----> {new_file}')

```

```python
city_names=['Cairo', 'Dhaka', 'Jakarta', 'Karachi', 'Manila',
 'Melbourne', 'Osaka', 'Shanghai', 'Singapore', 'Tokyo']
import shutil
for city in city_names:
    path=os.path.join('os-collaboration-exercise-data',city)
    os.mkdir(path)
    print(f'{city} folder is created')
    
    
    
```

```python
cairo_data=glob.glob('**/*cairo*',recursive=True)
for files in cairo_data:
    current_path=files
    target_path=os.path.join('.','os-collaboration-exercise-data','Cairo')
    shutil.copy(current_path,target_path)
    print(f'{files} is copied to Cairo folder')
```

```python
for city in city_names:
    data=glob.glob(f'**/*{city.lower()}*',recursive=True)
    target_path=os.path.join('.','os-collaboration-exercise-data',city)
    for files in data:
        destination_path=os.path.join(target_path,os.path.basename(files))
        if os.path.exists(destination_path):
            print(f'{os.path.basename(files)} already exists, skip move operation')
        else:
             shutil.move(files,target_path)
        print(f'{os.path.basename(files)} is moved to the target folder')
```
